```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.amazon.in/s?k=Home+Decor&crid=724FAEANTPA7&sprefix=home+deco%2Caps%2C262&ref=nb_sb_noss_2"
```

```python
headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}
```

```python
df = requests.get(url,headers=headers).text
```

```python
soup = BeautifulSoup(df,'html.parser')
```

```python
print(soup.prettify())
```

```python
links = soup.find_all("a",attrs={'class':'a-link-normal s-no-outline'})
```

```python
links[0].get('href')
```

```python
product = "https://amazon.in" + links[0].get('href')
```

```python
product
```

```python
title_list = []
title_list.clear()
rating_list = []
rating_list.clear()
price_list = []
price_list.clear()
brand_list=[]
brand_list.clear()
discount_list=[]
discount_list.clear()
rating_count=[]
rating_count.clear()
brought = []
brought.clear()
MSRP = []
MSRP.clear()
```

```python
for link in links:
    try:
       product = "https://amazon.in" + link.get('href')
       dt = requests.get(product,headers=headers).text
       new_soup = BeautifulSoup(dt,'html.parser')
    except Exception as e:
        print("Error in the starting....")
    try:
        
       title=new_soup.find("span",attrs={'class':'a-size-large product-title-word-break','id':'productTitle'}).text.strip()
       title_list.append(title)
    except Exception as e:
        print("Certain Error")
    try:
       span = new_soup.find_all("span",attrs={'class':'a-size-base a-color-base'})
       rate = span[1].text.strip()
       rating_list.append(rate)
    except Exception as e:
        print("Certain error in Rating fetching")
    try:
        price = new_soup.find_all("span",attrs={'class':'a-price-whole'})
        rate = price[0].text.strip()
        price_list.append(rate)
    except Exception as e:
        print("Price was not fetched ")
    try:
        brand = new_soup.find("a",attrs={'id':'sellerProfileTriggerId'}).text.strip()
        brand_list.append(brand)
    except Exception as e:
        print("brand was not able to fetch ")
    try:
        discount = new_soup.find("span",attrs={'class':'a-size-large a-color-price savingPriceOverride aok-align-center reinventPriceSavingsPercentageMargin savingsPercentage'}).text.strip()
        discount_list.append(discount)
    except Exception as e:
        print("Discount is the fetched ")
    try:
        r = new_soup.find("span",attrs={'id':'acrCustomerReviewText'}).text
        rating_count.append(r)
    except Exception as e:
        rating_count.append('NaN')
        print("Rating count for the product was not fetched ")
        
    try:
        b = new_soup.find_all("span",attrs={'class':'a-text-bold'})
        brought.append(b[9])
    except Exception as e:
        print("Brought count can't fetch")
    try:
        ms = new_soup.find_all("span",attrs={'class':'a-offscreen'})
        MSRP.append(ms[6])
    except Exception as e:
        print("MSRP can't be fetch")
```

```python
dict = {'title':title_list,'brand':brand_list,'rating_5.0':rating_list,'price':price_list,'discount':discount_list,'brought':brought,'rating':rating_count,'MSRP':MSRP}
```

```python
df = pd.DataFrame(dict)
```
